```python
import pandas as pd
import mysql.connector
from datetime import datetime

# Load data from CSV file (assuming claims_data.csv is still available)
claims_data = pd.read_csv('claims_data.csv')

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345",
    database="food_data"
)
cursor = conn.cursor()


# Function to convert timestamp format
def convert_timestamp(timestamp_str):
    """Convert MM/DD/YYYY HH:MM format to YYYY-MM-DD HH:MM:SS format"""
    try:
        # Parse the original format
        dt = datetime.strptime(timestamp_str, '%m/%d/%Y %H:%M')
        # Return in MySQL datetime format
        return dt.strftime('%Y-%m-%d %H:%M:%S')
    except ValueError:
        # If parsing fails, try alternative formats
        try:
            dt = datetime.strptime(timestamp_str, '%m/%d/%Y %H:%M:%S')
            return dt.strftime('%Y-%m-%d %H:%M:%S')
        except ValueError:
            print(f"Warning: Could not parse timestamp: {timestamp_str}")
            return None

# Convert timestamps in the DataFrame
claims_data['Timestamp'] = claims_data['Timestamp'].apply(convert_timestamp)

# Remove rows with invalid timestamps
claims_data = claims_data.dropna(subset=['Timestamp'])



# Insert data using iterrows()
try:
    for index, row in claims_data.iterrows():
        cursor.execute("""
            INSERT INTO claims_data (Claim_ID, Food_ID, Receiver_ID, Status, Timestamp)
            VALUES (%s, %s, %s, %s, %s)
        """, tuple(row))

    # Commit the transaction
    conn.commit()
    print(f"Successfully inserted {len(claims_data)} rows into claims_data!")

except Exception as e:
    print(f"Error inserting data: {e}")
    conn.rollback()

finally:
    cursor.close()
    conn.close()
```

## Food Providers and Receivers

```python
#1.How many food providers and receivers are there in each city
import mysql.connector
import pandas as pd

# Re-establish MySQL connection and cursor
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345",
    database="food_data"  # Specify the database
)
cursor = conn.cursor()

query="""
with base as (SELECT
    COALESCE(p.city, r.city) AS city,
    IFNULL(p.Providers_count, 0) AS Providers_count,
    IFNULL(r.receivers_count, 0) AS receivers_count
FROM
    (SELECT p.City AS city, COUNT(DISTINCT p.Provider_ID) AS Providers_count FROM providers_data p GROUP BY 1) AS p
LEFT JOIN
    (SELECT r.City AS city, COUNT(DISTINCT r.receiver_ID) AS receivers_count FROM receivers_data r GROUP BY 1) AS r
ON p.City = r.City

UNION

SELECT
    COALESCE(p.city, r.city) AS city,
    IFNULL(p.Providers_count, 0) AS Providers_count,
    IFNULL(r.receivers_count, 0) AS receivers_count
FROM
    (SELECT p.City AS city, COUNT(DISTINCT p.Provider_ID) AS Providers_count FROM providers_data p GROUP BY 1) AS p
RIGHT JOIN
    (SELECT r.City AS city, COUNT(DISTINCT r.receiver_ID) AS receivers_count FROM receivers_data r GROUP BY 1) AS r
ON p.City = r.City)

select distinct *
from base
order by Providers_Count desc

"""
cursor.execute(query)
result = cursor.fetchall()

df1 = pd.DataFrame(result, columns=["city","Providers_Count","Receivers_count"])
# Close the connection
cursor.close()
conn.close()
```
